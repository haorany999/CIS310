```python
d = {'col1':[1,4,3,6,8,3,4,6,5,9],
    'col2':[5,8,9,10,12,8,5,6,2,3]}

df2 = pd.DataFrame(d)
df2
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(df2['col1'],'k.',label='col1')
ax.plot(df2['col2'],'k--',label = 'col2')
ax.legend(loc='best')
```

```python
# Ticks, label, and legend.
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(randn(1000).cumsum())
```

```python
ticks = ax.set_xticks([0,250,500,750,1000])

labels = ax.set_xticklabels(['one','two','three','four','five'], rotation = 30, fontsize = 'small')

ax.set_title('My first matplotlip plot with ticks')

ax.set_xlabel('Stages')
ax.set_ylabel('Values')

fig
```

```python
# faster and easy way, using dictionary here
props = {'title':'My First Matplotlip Plot with Ticks',
         'xlabel':'Stages',
         'ylabel':'Vaules'}

ax.set(**props)
fig
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(randn(1000).cumsum(),'k',label= 'One')
ax.plot(randn(1000).cumsum(), 'k--',label = 'Two')
ax.plot(randn(1000).cumsum(), 'k.', label='Three')

ax.legend(loc='best')
```
