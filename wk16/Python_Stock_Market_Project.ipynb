```python
# fidning the constant columns in the data set.
constant_columns=[]
for c in data.columns:
    n = len(data[c].unique())
    if n==1 :
        constant_columns.append(c)
        
print('Constant Columns', constant_columns)
```

```python
# drop the constant colmns from the data set
data.drop(columns=constant_columns,axis=1, inplace=True)
data
```
